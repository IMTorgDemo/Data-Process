__Function Definitions:__ Specific definitions for distance will be used throughout the notebook.  Specifically, the [haversine model](https://en.wikipedia.org/wiki/Haversine_formula) calculates distance using lat,lon on a sphere.  I found similar algorithms in a few places, but the wikipedia page directed me to this [stackoverflow resource](https://stackoverflow.com/questions/4913349/haversine-formula-in-python-bearing-and-distance-between-two-gps-points/4913653#4913653).  I realize a wikipedia page and stackoverflow comment are far from a peer-reviewed publication standard.  However, after a quick, on-line literature review, these appear common-practice.  This could be improved with future work.

It can be argued that the haversine formula should not be used because most cities, such New York City, are laid-out in a grid.  However, Question 3 is mostly focused on obtaining walking distance to an airport.  As a pedestrian, the grid system becomes less important. 

If the haversine is seen as a gross approximation, Google Maps API may allow more accurate estimation.

```python
# calculate distance using model
from numpy import radians, sin, cos, arcsin, sqrt
# interesting discussion on performance, below
# ref: https://stackoverflow.com/questions/25767596/vectorised-haversine-formula-with-a-pandas-dataframe/25767765#25767765
def haversine(row, lon1, lat1):
    lon2 = row['Pickup_longitude']
    lat2 = row['Pickup_latitude']
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * arcsin(sqrt(a)) 
    km = 6367 * c
    return km
```

```python
# Google maps may be more accurate than model-based methods for calculating distance, as opposed to haversine
# I will use this platform if you would like to reimburse me :)
import requests, json 
  
def gmaps_api(source = "40.647491,-73.787827", dest = "Seattle, WA"):
    # returns distance (in km), duration (day hours)
    api_key = "<Google Maps API Key>"
    url ="https://maps.googleapis.com/maps/api/distancematrix/json?"
    r = requests.get(url + 'origins= ' + source +
                   '&destinations= ' + dest +
                   '&key= ' + api_key) 
    rslt = r.json()
    distance = rslt["rows"][0]["elements"][0]["distance"]["value"]
    duration = rslt["rows"][0]["elements"][0]["duration"]
    return { "distance":(distance,"km"), "duration":(duration["text"], duration["value"]) }
```

__Discussion on derived ratio variables:__ Creating a derived variable from other variables, especially when the calculation creates a ratio, should not be done without careful consideration.  There are many reasons for this, but the danger is mostly from not knowing the underlying mechanisms that created the original independent variables that are now mathematically combined.

Using ratios and percents in statistics has a long history (starting with Karl Pearson) of very passionate debate about their dangers, and it is often (conveniently) over-looked by general practicianers, as well as other scientific fields that make liberal use of ratios (econometrics, psychology, ...).  Even worse is that despite over a hundred years studying the topic, there still does not appear to be a general agreement for approaching this problem.  And there is certainly no consensus across disciplines.  Far from it, the subject is not broached often in classroom settings.

I performed a small literature review because I haven't looked at this subject in a while.  I found the following to be useful in how it pertains to linear models:
* Users Guide to Ratio Variables, Firebaugh and Gibbs; American Sociological Review (1985)
* Statistical Properties of Ratios, Atchley, Gaskins, and Anderson; Systematic Biology (1976)
* Explorations in statistics: the analysis of ratios and normalized data, Curran-Everett; Advances in Physiology Education (2013) 

Modeling with a ratio can have pitfalls.  Specifically, there could be some collinearity between dependent and independent variables that might have serious consequences for the assumptions of the model.  An interesting perspective of regressing on a ratio is suggested by [Kronmal 1993](http://www.jstor.org/stable/2983064).  Instead of using a dependent ratio variable, multiple each of the independent variables by the denominator.  However, this too has short-comings in that it increases collinearity among independent variables.

I will try to be cognizent of these concepts while working with the data.

__Discussion on the distribution of derived tip-percentage variable:__ We need to model tip as a percentage of the total fare.  Let us describe the situation, in detail, as it can be perceived in multiple ways.

The distribution of tip-percentage appears to be created through a highly complex process, which is made more difficult by the inherent biases of the dataset.  A large percentage of people (~13%) tip at the commonly-accepted rate of .20 of trip cost.  This leaves about 87% of tips which appear to be scattered from zero to 1 (or more), with decreasing occurence.

We can enumerate a few processes that might lead to this distribution:
1. getting a tip is a binary response, with little or no tip meanning failure and tip of .2 and higher being a success.
2. size of both tip amounts and total amounts are created through a random process, which are often modeled using the gamma distribution.  The ratio of sampling gamma random variables is the beta distribution.
3. there exists a mixture of the above two scenarios.

I am intrigued by the idea of modeling a complicated mixture of two processes; however, I feel diving into the details of scenarios 2 and 3 would be quite time-consuming.  Instead, scenario 1 is straight-forward and a good initial approach.  It simplifies the problem of biases in the dataset, as well as probles associated with a ratio dependent variable.  Because so many people simply tip at the accepted-rate, this rule divides the dataset in half.  With the dataset already neatly partitioned, setting the dependent variable to success / failure is not an unreasonable assumption.

Logistic regression is a popular tool for modeling binary response.

__Datasets for training and testing:__ Because the data is unbalanced, we will create a balanced dataset by sampling with replacement on the two categories.  While predictive analysis is not as sensitive to unbalanced data as generative analysis, it can still cause difficulties reading a confusion matrix or ROC curve.  We will only use a subset of the data because of limited machine memory.  If we want to make use of all the data, then we can run a spark job, later.

```python
# lets create a balanced sample, so that we can more easily compare the two groups
notip = data.loc[data.Tip_pct_bin==0,].sample(n=25000, replace=True, random_state=1)
tip = data.loc[data.Tip_pct_bin==1,].sample(n=25000, replace=True, random_state=1)
smpl = tip.append(notip)
```

```python
# Data Preparation
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

numeric = smpl[[ 'Passenger_count', 'Trip_distance']]
categoric = smpl[[ 'AirInd', 'RateCode', 'TripType', 'PickUpHr']]

lb = LabelEncoder()
categoric["AirIndEnc"] = lb.fit_transform(smpl.AirInd)
categoric["RateEnc"] = lb.fit_transform(smpl.RateCode)
categoric["TripTypeEnc"] = lb.fit_transform(smpl.TripType)
categoric["PickUpHrEnc"] = lb.fit_transform(smpl.PickUpHr)
#lb.inverse_transform(y) #get back to original

encoded = categoric[[ 'AirIndEnc', 'RateEnc', 'TripTypeEnc', "PickUpHrEnc"]]

df = pd.concat([encoded, numeric], axis=1)
y = np.ravel( smpl[[ 'Tip_pct_bin' ]] )


# Create training and testing data
PCT_TEST = .2
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=PCT_TEST)

print( X_train.shape, y_train.shape )
print( X_test.shape, y_test.shape )
```

__Modeling methodology:__ The following is our model validation, training, and testing methodology:
1. create Train,Test with split
2. use k-folds CV with Training set on multiple model-families (unparamterized)
3. select the best model-family
4. apply Training set to model-familty to parameterize it
5. use Testing set to evaluate final model characteristics

_Note:_ Train and Test should be from different (independent) data sets, if possible.

__Model:__ The logistic regression (supervised classification) model can be applied to binary response data in a predictive scenario.  It is characterized as a generalized linear model with the logit link function 'Categorical Data Analysis', Agresti (2002).  The logistic function is a sigmoid curve that can be written as the following:

\begin{eqnarray}
 p(x) = \operatorname{Pr}(\text{Tip} = 1 ) = \frac{\exp(\beta_{0} + \beta_{1}x) }{1 + \exp(\beta_{0} + \beta_{1}x)} \label{eq:glm1} 
\end{eqnarray}

By taking the inverse of the logistic, we define the logit function, g, or log-odds, which expresses it as a linear regression:


\begin{eqnarray}
{\displaystyle g(p(x))=\operatorname {logit} p(x)=\ln \left({\frac {p(x)}{1-p(x)}}\right)=\beta _{0}+\beta _{1}x,}
\end{eqnarray}

While interpreting log-odds with coefficients can be tricky, this is less of a concern in a predivtive context because of the emphasis on the outcome.

<br>

We can use this model for predicting the outcome of a taxi driver getting a tip because the vast majority (~90%) of riders either give little to no tip, or the popularly-accepted tip of around 20%.  The algorithm provided by the [sklearn module](http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html) includes many helpful functions to support a nice modeling process. 

Because we do not have many predictors, and the specific sklearn logistic regression allows for L1 or L2 regularization, there is not much need for dimensionality reduction techniques, such as Principal Components Analysis or MultiFactor Analysis, before training the model.  The fact that the model allows for L1 regularization is often referred to as L1-based feature selection.  This also means we do not have to worry about normalizing and standardizing our independent variables, using the default solver (liblinear), which is a necessary step for those analyses.

A more advanced model that sklearn proposes is LogisticRegressionCV.  This has a built-in cross validation implementation that allows it to automatically tune hyperparameters, specifically the C constant hyperparameter in-front of the cost function's sum of logs.  Here, smaller C specifies stronger regularization.  Manually, we could use a grid search to tune the model for the hyperparameter.  It might be interesting to train both models, then compare their performance with the test data to see if tuning the C parameter has an effect.

We will also try a [DecisionTree classifier](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html) with hyperparamter-tuning performed manually using grid search.  This is only for comparison purposes so that we have an idea how another model looks against logistic regression.

<br>

```python
# CV: LogisticRegression
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression(penalty='l1',solver='liblinear')

# Validate selected model
from sklearn import metrics, model_selection
scores = model_selection.cross_val_score(logreg, X_train, y_train, cv=10)          # use default scoring='accuracy'
print(scores)
print('average score: {}'.format(scores.mean()))
```

```python
# Train model
logreg.fit(X_train, y_train)
```

```python
# Test model
predicted = logreg.predict(X_test)              # predict class labels for the test set
probs = logreg.predict_proba(X_test)            # p(x) = 1/[1+exp(-BetaX)] generate class probabilities - (soft) classifier outputting the probability of the instance being in each of the classes.
y_score = logreg.decision_function(X_test)      # d(x) = BetaX distance to the separating hyperplane, or however the classifier evaluates (universal)

# generate evaluation metrics
print( metrics.accuracy_score(y_test, predicted) )
print( metrics.roc_auc_score(y_test, probs[:, 1]))
```

```python
# CV: tunes hyper-parameter-C with built-in cross-validation
from sklearn.linear_model import LogisticRegressionCV

# Train
clf = LogisticRegressionCV(cv=5, random_state=0, multi_class='ovr')
clf.fit(X_train, y_train)
```

```python
# Test
#X.iloc[:2, :]                         #data recs
#clf.predict(X.iloc[:2, :])            #Predict class labels for samples in X
#clf.predict_proba(X.iloc[:2, :])      #Probability estimates for each category, for each input record
clf.score(X_test, y_test)              #accuracy of the model based on the test dataset
```

```python
# CV: Decision Tree
from sklearn import tree
from sklearn.model_selection import GridSearchCV

parameters = {'max_depth':range(2,20)}
clfTree = GridSearchCV(tree.DecisionTreeClassifier(), parameters, n_jobs=4, cv=10)
clfTree.fit(X=X_train, y=y_train)
treeModel = clfTree.best_estimator_
print (clfTree.best_score_, clfTree.best_params_) 
```

```python
# Test model
predicted = treeModel.predict(X_test)              # predict class labels for the test set
probs = treeModel.predict_proba(X_test)            # p(x) = 1/[1+exp(-BetaX)] generate class probabilities - (soft) classifier outputting the probability of the instance being in each of the classes.
#y_score = treeModel.decision_function(X_test)      # d(x) = BetaX distance to the separating hyperplane, or however the classifier evaluates (universal)

# generate evaluation metrics
print( metrics.accuracy_score(y_test, predicted) )
print( metrics.roc_auc_score(y_test, probs[:, 1]))
```

So, we leave the model as simply determined by hour-of-day and week. 

Analysis Of Variance (ANOVA) assumes normality and homoscedasticity.  These assumptions are not strongly supported, here; however, ANOVA may be a good initial test.  More robust methods can be attempted, in future work.

The ANOVA tests maintain the previous results and confirm the suspicions of the box and whisker plot: we reject the null that the means among hours are equal.  When we look at the t-tests, many of the hours reject the null that they have zero effect, individually.  Future work may include fitting a non-linear model to the changing average speed, thoughout the day.
